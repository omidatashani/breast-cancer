```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_breast_cancer
```

```python
# Loading dataset
cancer = load_breast_cancer()
# Making pandas dataframe
df = pd.DataFrame(cancer.data, columns=cancer.feature_names)
df['target'] = pd.Series(cancer.target)
df.head(20)
```

```python
X = df.drop('target', axis=1)
X.head()
```

```python
Y = df.T.tail(1).T
Y.head()
```

```python
# Check for missing values
df.isnull().sum()
```

```python
# Show data type of each feature
df.info()
```

```python
# Count the occurrences of cancer and no cancer
df['target'].value_counts()
```

```python
# Plot histograms for each variable
df.hist(figsize=(10, 10))
plt.tight_layout()
plt.show()
```

```python
# Visualize the correlation matrix
plt.figure(figsize=(30,30))
sns.heatmap(df.corr(), cmap='coolwarm', annot=True, square=True)
plt.show()
```

```python
# Pairplot of selected features
selected_features = ['mean radius', 'mean texture', 'mean perimeter', 'mean area', 'mean smoothness', 'mean compactness']
sns.pairplot(df, hue='target', vars = selected_features)
plt.show()
```

```python
# Scatter plot
sns.scatterplot(x='mean radius', y='mean texture', hue='target', data=df)
```

```python
# Boxplot of selected features
plt.figure(figsize=(20, 10))
for i, feature in enumerate(selected_features):
    plt.subplot(1, len(selected_features), i+1)
    sns.boxplot(data=df[feature], orient="h", palette="Set2")
plt.show()
```

```python
# Violin plots of selected features grouped by the target variable
plt.figure(figsize=(20, 10))
for i, feature in enumerate(selected_features):
    plt.subplot(1, len(selected_features), i+1)
    sns.violinplot(x="target", y=feature, data=df)
plt.tight_layout()
plt.show()
```
